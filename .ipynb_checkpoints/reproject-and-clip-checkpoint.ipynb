# Import Data 

````python
# states from US census
states = gpd.read_file('data/tl_2022_us_state/tl_2022_us_state.shp')
states.columns = states.columns.str.lower()

# populated places from Natural Earth 
places = gpd.read_file('data/ne_50m_populated_places_simple/ne_50m_populated_places_simple.shp')

roads = gpd.read_file('data/ne_10m_roads/ne_10m_roads.shp')
````

We want to make a map of Alaska. Lets select it!

````python
alaska = states[states.name == 'Alaska']

alaska.plot()
````

## Reprojecting

issue: Alaksa's islands are elongating the map

fix: reproject the alaska geo-datafrmae

**Reprojecting**: change the crs of geospatial data

Reproject to crs EPSG:3338 for our purposes.

Optimizes data working with data from Alaska.

Use `to_crs()` method for `gpd.GeoDataFrame`s:

Syntax:
```python

updated_geodf = geodf.to_crs(new_crs)

```

where...

- `updated_geodf` is reprojected geodataframe - not in place! Making new one
- `geodf` is geodataaframe you wnat to repreoject
- `new_crs` the crs we want to convert to, and object of type crs of a string representing crs (eg. 'epsg:3338')

````python
# change to projected
alaska = alaska.to_crs('epsg:3338')

alaska.plot()
````

````python
# check info 
print('is the crs projected?', alaska.crs.is_projected)
````

## Prepare populated places

### Exploration

````python
print(places.crs)
places.head(3)

places.plot()
````

`places` has. crs epsg:4326, which is forg WGS84. Most common for global data. 

What happens if we try to plot `places` and `alaska` together? These have different crs, so they will not interact.

````python
# classic bafoonery! We cant do this! They have different crs
fig, ax = plt.subplots()

alaska.plot(ax = ax)
places.plot(ax = ax, color = 'pink')

plt.show()
````

````python
# these are different
print(alaska.crs)
print(places.crs)
````

### Reproject places data. 
Use `to_crs` to reproject...again!

````python
# specify the alaska crs for reproducibility
places = places.to_crs(alaska.crs)

print(alaska.crs == places.crs)
````

````python
# plotted all of the places data
fig, ax = plt.subplots()

alaska.plot(ax = ax)
places.plot(ax = ax, 
            color = 'pink',
           markersize = 4)

plt.show()
````

### Clipping

**Clipping** means using a polygon to only select geospatial data within those polygons.

To clip `gpd.GeoDataFrame` we can use geopandas `clip()` function.

Syntax:
```python

updated_geodf = geodf.clip(geodf, mask)

```

where...

- `updated_geodf` output of function, the intersection fo geometries in `geodf` with `mask`
- `geodf` is geodataaframe we want to clip
- `mask` is a gedataframe with polygons we want to use for clipping. Has to have the SAME crs as `geodf`

````python
# use alaska as the mask because we only want places there
ak_places = gpd.clip(places, alaska)
ak_places
````

````python
# plotted all of the places data
fig, ax = plt.subplots()

alaska.plot(ax = ax)
ak_places.plot(ax = ax, 
            color = 'violet',
           markersize = 10)

plt.show()
````

### Data Selection

Geospatial operations use a lot of resources, so its good to reduce data before modifying it. 

````python
# select only usa roads

usa_roads = roads[roads.sov_a3 == 'USA']
usa_roads.plot()
````

### Clipping with the bounding box

Clip the `usa_roads` with the bounding box of `alaska` instead of its polygons.

To create box, we use the `box()` function we imported from `shapely.geometry`.

Syntax:
```python
box(minx, miny, maxx, maxy)

```

If we want to create a bounding box from the bounds of a geodataframe `gdf` we can use this syntax:
```python
box(*gdf.total_bounds)

```


````python
# to get the bounding box of this gdf
## output is array with minx, miny, maxx, maxy coordinates
alaska.total_bounds
````

````python
# the * is being used an an unpacking operator 
## to unpack the contents of the alaska.total_bounds 
## and assigns them as parameters to box
bbox = box(*alaska.total_bounds)
bbox
````

````python
type(bbox)
````

````python
# make alaska bounding box
ak_bbox = gpd.GeoDataFrame( geometry = [bbox],
                          crs = alaska.crs)

ak_bbox
````

````python
# repreoject and clip
ak_roads = gpd.clip(usa_roads.to_crs(ak_bbox.crs), # reprojecting usa_roads
                    ak_bbox) # clipping it to bounding box
````

````python
# plotted all of the places data
fig, ax = plt.subplots(figsize = (12,8))

# remove axes
ax.axis('off')

alaska.plot(ax = ax,
           color = 'none',
            edgecolor = '0.7')
ak_places.plot(ax = ax, 
            color = 'violet',
           markersize = 20,
              marker = 's') # square markers
ak_roads.plot(ax = ax,
         color = 'red')

plt.show()
````
