## About our data

Dataset 1:
- from the US Energy Information Administration (EIA). 
- info about current operable electric power plants in the US by energy source 

Dataset 2: TIGER shapefules from us census bureau depicting state boudnaries 

```python
# import necessary libraries

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# update pandas display option 
pd.set_option("display.max.columns", None)
```

We want to make a geopandas geodatafrmae from csv data

```python
# import csv data using pandas
power_plants = pd.read_csv("data/power_plants_epsg4269.csv")
```

From documentation we know that CRS is EPSF:4269 in lat/long coordinates.

Use this info to create new `gpd.GeoDataFrame` from `pd.DataFrame`

To do this we use `gpd.points_from_xy` function:

```python
# overwrite name and read in power plants data frame
power_plants = gpd.GeoDataFrame(power_plants, # data for geo-dataframe
                               # specify a geomoetry column 
                               geometry = gpd.points_from_xy(power_plants.longitude,
                                                            power_plants.latitude),
                               # specify CRS
                               crs= 'EPSG:4269'
                               )

power_plants.head()
```

```python
# print CRS info
print("is geographic?", power_plants.crs.is_geographic)
print("is projected?", power_plants.crs.is_projected)
print("datum?", power_plants.crs.datum)
print("ellipsoid?", power_plants.crs.ellipsoid)
```

```python
# read in the Tiger shapefiles
states = gpd.read_file('data/tl_2022_us_state/tl_2022_us_state.shp')

states.head()
```

```python
# simplify columns to lower case
states.columns = states.columns.str.lower()

# check dtypes of columns
states.info()

# reassign these regions to be type integer
states[['region', 'division', 'statefp']] = states[['region', 'division', 'statefp']].astype('int')
```

```python
# check to see if they changed to integer
states.info()
```

```python
# check crs; EPSG:4269; NAD83
states.crs

# false
states.crs.is_projected

# true - it is geographic 
states.crs.is_geographic

# the crs of these does match
states.crs == power_plants.crs
```

```python
# plot
# initialize an empty figure 
fig, ax = plt.subplots()

# add data to axis 
states.plot(ax = ax,
            alpha = 0.2,
            color = 'blue')

# the min and max of the data goes from -180 to 180 degrees
```

```python
states[['statefp', 'name']]

# fp codes 78, 69, 66, 60, 72 are not US states
```

```python
# make list of non states use statefp
non_states = states[(states['statefp'] == 2) | (states['statefp'] == 78) | (states['statefp'] == 69) | (states['statefp'] == 66) | (states['statefp'] == 60) | (states['statefp'] == 72)].statefp

non_states.index

states_cont = states.drop(non_states.index)
```

```python
# remove regions that are not state > 4
cont = states[(states['region'] != 9) & (~states.statefp.isin([2, 15]))]

cont.info()
```

```python
power_plants = power_plants[~power_plants.state.isin(['Puerto Rico', 'Hawaii', 'Alaska'])]

# check crs
power_plants.crs == cont.crs
```
