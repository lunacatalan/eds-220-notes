# `rioxarray`


`rioxarray`:
- this is a python extenstion for `xarray` to manipualte ``xarray.DataArray` as rasters
- means raster input/output + xarray

## Data
Raster files from the US National Agriculture Imagery Program (NAIP)

- high-resolution aerial images from 2020 with four spectral bands: Red, Green, Blue and Near-Infrared (NIR)

# Import .tif

```python
import os # to work with filepaths
import numpy as np
import matplotlib.pyplot as plt

import rioxarray as rioxr

import geopandas as gpd
from shapely.geometry import Polygon
```

There are multiple ways to open .tif files using `xarray` or `rioxarray`.

Use `rioxarray.open_rasterio()` function to open the .tif file:

```python
# get current working directory
os.getcwd()
```

```python
# load NIR .tif file 

#create filepath
nir_fp = os.path.join(os.getcwd(), 'data', 'nir.tif')

# open file
nir = rioxr.open_rasterio(nir_fp)
nir
```

## `rio`

An **accessor** in Python lets us use a different set of propoerties from an object. 

Use the `.rio` accessor for `xarray.DataArray` to access its raster properties. 

**Example**

```python
print("# bands:", nir.rio.count)
print("height:", nir.rio.height)
print("width:", nir.rio.width)
print("resolution:", nir.rio.resolution())
print("no data values:", nir.rio.nodata, '\n') # the '\n' just makes a space when printing

print('spatial bounding box:')
print(nir.rio.bounds(), '\n')

print('CRS:', nir.rio.crs)
```

```python
# open RGB raster
rgb_fp = os.path.join(os.getcwd(), 'data', 'rgb.tif')
rgb = rioxr.open_rasterio(rgb_fp)
rgb
```

```python
# check some number bands 
print('# bands:', rgb.rio.count)
```

```python
# parameters for plotting this raster

size = 6 # height of figure in inch
aspect = rgb.rio.width/rgb.rio.height # ratio of w to h to plot proportionally 

# plot 3 bands as RGB image
rgb.plot.imshow(size = size,
               aspect = aspect)
```

## Box for clipping 

Our area of interest (aoi) is a smaller region with only a few blocks around the NCEAS building 

```python
# list of points

points = [
    [
              -119.70325533542601,
              34.425768833663085
            ],
            [
              -119.70325533542601,
              34.41869601583825
            ],
            [
              -119.68994975778048,
              34.41869601583825
            ],
            [
              -119.68994975778048,
              34.425768833663085
            ],
            [
              -119.70325533542601,
              34.425768833663085
            ]
          ]
```

```python
# create a geodataframe with aoi
aoi = gpd.GeoDataFrame(geometry = [Polygon(points)], #create polygon as geometry column
                      crs = 'epsg:4326')

aoi
```

```python
aoi.plot()
```

## Clip raster

Remember: for two geospatial sets to interact they need to have the same crs

Check:

```python
print('aoi CRS:', aoi.crs)
print('nir CRS:', nir.rio.crs)
print('rgb CRS:', rgb.rio.crs)
```

```python
# reproject aoi to crs
aoi = aoi.to_crs(nir.rio.crs)
print('matched crs?', aoi.crs == nir.rio.crs)
aoi.crs
```

```python
# plot aoi outline
fig, ax = plt.subplots()
fig.set_size_inches((size, size*aspect)) 

rgb.plot.imshow(ax = ax)
aoi.plot(ax = ax,
        alpha = 0.5)
```

To clip the raster using the aoi polygon we use the `.rio.clip_box()` method:

```python
# clip rasters to aoi
rgb_small = rgb.rio.clip_box(*aoi.total_bounds)
nir_small = nir.rio.clip_box(*aoi.total_bounds)
```

```python
print('original shape:', rgb.shape)
print('original shape:', rgb_small.shape)

# made smaller
```

```python
# plot with correct ratio
size = 4 # height in inches
aspect = rgb_small.rio.width/rgb_small.rio.height
rgb_small.plot.imshow(size = size, aspect = aspect)
```
